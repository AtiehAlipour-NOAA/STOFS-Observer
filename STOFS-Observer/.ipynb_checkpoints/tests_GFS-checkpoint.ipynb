```python
# Define parameters
start_date = '20240922'
end_date = '20240923'
cycles = ['00', '06', '12', '18']
num_time_steps = 6  # Number of time steps to retrieve

# Create a DataFrame for wind stations (example data)
stations = pd.DataFrame({
    'lat': [30.0, 31.0],  # Example latitudes
    'lon': [-80.0, -81.0],  # Example longitudes
    'nos_id': [123, 456]  # Example IDs
})

# Fetch GFS data
gfs_wind_u_data,gfs_wind_v_data, gfs_surface_pressure_data, time  = fetch_gfs_Nowcast_data(start_date, end_date, cycles, stations, num_time_steps)

print(gfs_wind_u_data)
print(gfs_wind_v_data)
print(gfs_surface_pressure_data)
print(time)

```
